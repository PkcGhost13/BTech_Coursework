```python
import numpy as np
import pandas as pd
df=pd.read_csv("studentsmarks1.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
quiz_avg=[]
for i in range(1,df.shape[1]-1):
    print("Avg score of quiz ",i," ",np.average(df.iloc[:,i].values))
    quiz_avg.append(np.average(df.iloc[:,i].values))
```

```python
avg_score=[]
for i in range(df.shape[0]):
    print("Avg score of student ",i,"\t",np.average(df.iloc[i,1:].values))
    avg_score.append(float(np.average(df.iloc[i,1:].values)))
```

```python
print("The quiz with the highest average marks is : Quiz -",quiz_avg.index(max(quiz_avg))+1,"with the average score of : ",max(quiz_avg))
```

```python
sorted_list = sorted(enumerate(avg_score,1), key=lambda x: x[1], reverse=True)
top_20 = sorted_list[:20]
print(top_20)
```

```python
count=0
for i in avg_score:
    if i>=60:
        count+=1
print("The number of students who passed are :",count)
```
