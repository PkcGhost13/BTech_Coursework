```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 100)
pd.set_option('display.colheader_justify', 'left')
pd.set_option('display.precision', 5)
```

```python
df = pd.read_excel("10 Districtwise Missing Persons_2021.xlsx")
df.iloc[0:3]
```

```python
df1=df.loc[df['Districtwise Missing Persons - 2020'] == 'State: West Bengal'].dropna()
df1['Unnamed: 2'] = df1['Unnamed: 2'].astype(int) 
s ='Unnamed: '
for i in range (2,25):
    y = str(i)
    e = s + y
    df1[e] = df1[e].astype(int)
df1
```

```python
#Children
temp=df1.iloc[0:39]
df2 = temp.sort_values('Unnamed: 24', ascending=False)
df2 = df2.iloc[0:10]
sns.set(rc={'figure.figsize': (30, 7)})
ax = sns.barplot(x=df2['Unnamed: 1'],y=df2['Unnamed: 24'])
ax.set_xlabel('District', fontsize=20)
ax.set_ylabel('Number of Children missing', fontsize=20)
list_murder = set(df2['Unnamed: 1'])
```

```python
child_crime = df1.iloc[0:39]['Unnamed: 3'].sum()
twelve_crime = df1.iloc[0:39]['Unnamed: 4'].sum()
sixteen_crime = df1.iloc[0:39]['Unnamed: 5'].sum()
adult_crime = df1.iloc[0:39]['Unnamed: 7'].sum()
values = [child_crime,twelve_crime,sixteen_crime,adult_crime]
labels = ['Male_Below 12 years', 'Male between 12 to 16','Male between 16 to 18','Male above 18']
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Missing Male')
plt.show()
```

```python
child_crime = df1.iloc[0:39]['Unnamed: 9'].sum()
twelve_crime = df1.iloc[0:39]['Unnamed: 10'].sum()
sixteen_crime = df1.iloc[0:39]['Unnamed: 11'].sum()
adult_crime = df1.iloc[0:39]['Unnamed: 13'].sum()
values = [child_crime,twelve_crime,sixteen_crime,adult_crime]
labels = ['Female_Below 12 years', 'Female between 12 to 16','Female between 16 to 18','Female above 18']
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Missing Female')
plt.show()
```

```python
male_crime = df1.iloc[0:39]['Unnamed: 2'].sum()
female_crime = df1.iloc[0:39]['Unnamed: 8'].sum()
trans_crime = df1.iloc[0:39]['Unnamed: 14'].sum()
values = [male_crime,female_crime,trans_crime]
labels = ['Male missing', 'Female missing','Trans Missing']
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Missing Male,Female and trans people', fontsize=17)
plt.show()
```

```python
plt.plot(df1.iloc[0:39]['Unnamed: 1'], df1.iloc[0:39]['Unnamed: 7'], marker='o',label='Missing Male')
plt.plot(df1.iloc[0:39]['Unnamed: 1'], df1.iloc[0:39]['Unnamed: 13'],marker='o',label='Missing Female')
plt.xlabel("Districts", fontsize=20)
plt.ylabel("Missing Adult Male ", fontsize=20)
plt.title("Missing Adult Female", fontsize=20)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
cols = ['Unnamed: 7', 'Unnamed: 13']
plt.figure(figsize=(6,3))
dataplot = sns.heatmap(df1[cols].corr(), annot=True)
ax = dataplot.axes
ax.set_xticklabels(['Number of Missing Male', 'Number of Missing Female'], rotation=45)
ax.set_yticklabels(['Number of Missing Male', 'Number of Missing Female'], rotation=0)
plt.show()
```
