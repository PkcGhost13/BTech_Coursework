```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))
axs[0].scatter(df.index, df['sepal.width'], c='green')
axs[0].set_title('Sepal Width')
axs[1].scatter(df.index, df['sepal.length'], c='blue')
axs[1].set_title('Sepal Length')
axs[2].scatter(df.index, df['petal.width'], c='yellow')
axs[2].set_title('Petal Width')
axs[3].scatter(df.index, df['petal.length'], c='red')
axs[3].set_title('Petal Length')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15,5))

sns.scatterplot(x=df.index, y=df['sepal.width'], hue=df['variety'], ax=axs[0])
sns.scatterplot(x=df.index, y=df['sepal.length'], hue=df['variety'], ax=axs[1])
sns.scatterplot(x=df.index, y=df['petal.width'], hue=df['variety'], ax=axs[2])
sns.scatterplot(x=df.index, y=df['petal.length'], hue=df['variety'], ax=axs[3])

plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15,5))

axs[0].set_title('Line plot of sepal length')
axs[0].set_xlabel('index')
axs[0].set_ylabel('sepal length')
axs[0].plot(df.index, df['sepal.length'], markevery=1, marker='d')
for name, group in df.groupby('variety'):
    axs[0].plot(group.index, group['sepal.length'], label=name, markevery=1, marker='d')
axs[0].legend()

axs[1].set_title('Line plot of sepal width')
axs[1].set_xlabel('index')
axs[1].set_ylabel('sepal width')
axs[1].plot(df.index, df['sepal.width'], markevery=1, marker='d')
for name, group in df.groupby('variety'):
    axs[1].plot(group.index, group['sepal.width'], label=name, markevery=1, marker='d')
axs[1].legend()

axs[2].set_title('Line plot of petal length')
axs[2].set_xlabel('index')
axs[2].set_ylabel('petal length')
axs[2].plot(df.index, df['petal.length'], markevery=1, marker='d')
for name, group in df.groupby('variety'):
    axs[2].plot(group.index, group['petal.length'], label=name, markevery=1, marker='d')
axs[2].legend()

axs[3].set_title('Line plot of petal width')
axs[3].set_xlabel('index')
axs[3].set_ylabel('petal width')
axs[3].plot(df.index, df['petal.width'], markevery=1, marker='d')
for name, group in df.groupby('variety'):
    axs[3].plot(group.index, group['petal.width'], label=name, markevery=1, marker='d')
axs[3].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15,5))

sns.stripplot(y=df['sepal.width'],ax=axs[0], hue=df['variety'])
sns.stripplot(y=df['sepal.length'],ax=axs[1], hue=df['variety'])
sns.stripplot(y=df['petal.width'],ax=axs[2], hue=df['variety'])
sns.stripplot(y=df['petal.length'],ax=axs[3], hue=df['variety'])

plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15,5))

sns.stripplot(x=df['variety'],y=df['sepal.width'], ax=axs[0],hue=df['variety'])
sns.stripplot(x=df['variety'],y=df['sepal.length'],ax=axs[1], hue=df['variety'])
sns.stripplot(x=df['variety'],y=df['petal.width'],ax=axs[2], hue=df['variety'])
sns.stripplot(x=df['variety'],y=df['petal.length'],ax=axs[3], hue=df['variety'])

plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15,5))

sns.swarmplot(x=df['sepal.width'],ax=axs[0],hue=df['variety'])
sns.swarmplot(x=df['sepal.length'],ax=axs[1],hue=df['variety'])
sns.swarmplot(x=df['petal.width'],ax=axs[2],hue=df['variety'])
sns.swarmplot(x=df['petal.length'],ax=axs[3],hue=df['variety'])

plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15, 5))

axs[0].hist(df['sepal.width'], color='red')
axs[0].set_title('Sepal Width')

axs[1].hist(df['sepal.length'], color='green')
axs[1].set_title('Sepal Length')

axs[2].hist(df['petal.width'], color='yellow')
axs[2].set_title('Petal Width')

axs[3].hist(df['petal.length'], color='blue')
axs[3].set_title('Petal Length')

plt.show()
```

```python
fig, axs = plt.subplots(ncols=4, figsize=(15, 5))

df['sepal.length'].plot(kind='density', ax=axs[0], title='Sepal Length')
df['sepal.width'].plot(kind='density', ax=axs[1], title='Sepal Width')
df['petal.length'].plot(kind='density', ax=axs[2], title='Petal Length')
df['petal.width'].plot(kind='density', ax=axs[3], title='Petal Width')

plt.show()
```
