```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\rajhe\\Downloads\\iris.csv")
print(df.describe())
plt.scatter(df.index,df['sepal.length'],color='blue') ; plt.xlabel('sepal.length') ; plt.ylabel('Frequency')
plt.show()
plt.scatter(df.index,df['sepal.width'],color='red'); plt.xlabel('sepal.width') ; plt.ylabel('Frequency')
plt.show()
plt.scatter(df.index,df['petal.length'],color='green'); plt.xlabel('petal.length') ; plt.ylabel('Frequency')
plt.show()
plt.scatter(df.index,df['petal.width'],color='orange'); plt.xlabel('petal.width') ; plt.ylabel('Frequency')
plt.show()

```

```python
sns.scatterplot(x=df.index,y=df['sepal.length'],hue=df['variety'])
plt.show()
sns.scatterplot(x=df.index,y=df['sepal.width'],hue=df['variety'])
plt.show()
sns.scatterplot(x=df.index,y=df['petal.length'],hue=df['variety'])
plt.show()
sns.scatterplot(x=df.index,y=df['petal.width'],hue=df['variety'])
plt.show()
```

```python
plt.title('Line plot of sepal length')
plt.xlabel('index')
plt.ylabel('sepal length')
plt.plot(df.index, df['sepal.length'],markevery=1,marker='d')
for name, group in df.groupby('variety'):
    plt.plot(group.index,group['sepal.length'],label=name,markevery=1,marker='d')
plt.legend()
plt.show()

plt.title('Line plot of sepal width')
plt.xlabel('index')
plt.ylabel('sepal width')
plt.plot(df.index, df['sepal.width'],markevery=1,marker='d')
for name, group in df.groupby('variety'):
    plt.plot(group.index,group['sepal.width'],label=name,markevery=1,marker='d')
plt.legend()
plt.show()

plt.title('Line plot of petal length')
plt.xlabel('index')
plt.ylabel('petal length')
plt.plot(df.index, df['petal.length'],markevery=1,marker='d')
for name, group in df.groupby('variety'):
    plt.plot(group.index,group['petal.length'],label=name,markevery=1,marker='d')
plt.legend()
plt.show()

plt.title('Line plot of petal width')
plt.xlabel('index')
plt.ylabel('petal width')
plt.plot(df.index, df['petal.width'],markevery=1,marker='d')
for name, group in df.groupby('variety'):
    plt.plot(group.index,group['petal.width'],label=name,markevery=1,marker='d')
plt.legend()
plt.show()
```
